```python
import pandas as pd

# 1. LOAD DATA
df = pd.read_csv('zomato.csv')

print("=== BEFORE CLEANING ===")
print("Total rows:", len(df))

# 2. CLEANING
# Remove # @ ' and dots
df['restaurant name'] = df['restaurant name'].str.replace('#', '', regex=False)
df['restaurant name'] = df['restaurant name'].str.replace('@', '', regex=False)
df['restaurant name'] = df['restaurant name'].str.replace("'", '', regex=False)
df['restaurant name'] = df['restaurant name'].str.replace('.', '', regex=False)
df['restaurant name'] = df['restaurant name'].str.strip()

# Remove rows with garbage
df = df[~df['restaurant name'].str.contains('AAAA|B.C|°C', na=False)]

# Remove rows with weird foreign characters
df = df[df['restaurant name'].apply(lambda x: x.isascii() if pd.notnull(x) else False)]

# Remove duplicates + nulls
df = df.drop_duplicates(subset=['restaurant name'])
df = df.dropna()

# Fix numbers
df['rate (out of 5)'] = pd.to_numeric(df['rate (out of 5)'], errors='coerce')
df['avg cost (two people)'] = pd.to_numeric(df['avg cost (two people)'], errors='coerce')
df = df[df['rate (out of 5)'] <= 5]
df = df[df['avg cost (two people)'] < 5000]

# 3. SAVE CLEANED FILE
df.to_csv('cleaned_dataset.csv', index=False)

print("\n=== AFTER CLEANING ===")
print("Total rows:", len(df))
print("Rows removed:", 7105 - len(df))

# 4. CHECK IF IT'S REALLY CLEAN
print("\n=== CHECK FOR BAD CHARACTERS ===")
bad_chars = ['#', '@', "'", '.']
for char in bad_chars:
    count = df['restaurant name'].str.contains(char, na=False).sum()
    print(f"Character '{char}' found in {count} rows - should be 0 ✓")

print("\n=== 10 CLEAN SAMPLE NAMES ===")
print(df['restaurant name'].head(10).to_list())
```
